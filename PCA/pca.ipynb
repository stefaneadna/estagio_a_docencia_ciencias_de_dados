## Importa as bibliotecas

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

## Carregando a base de dados Iris

```python
iris = load_iris()
X = iris.data
```

```python
X
```

## Normalização dos dados

```python
X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
```

## Cálculo do PCA

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_norm)
X_pca
```

## Variância explicada pelos componentes principais

```python
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_ratio
```

## Visualização dos resultados

```python
print("Variância explicada pelos componentes principais:", explained_variance_ratio)
print("Componentes principais:\n", pca.components_)
```

## Utilizar os dados para classificar

```python
y = iris.target
# Dividir os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Treinar o classificador (regressão logística)
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

# Avaliar o classificador
accuracy = classifier.score(X_test, y_test)
print("Acurácia do classificador: ", accuracy)
```
